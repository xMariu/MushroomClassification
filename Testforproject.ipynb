```python
df = df.astype('category')
df.dtypes
```

```python
le=preprocessing.LabelEncoder()
for column in df.columns:
    df[column] = le.fit_transform(df[column])
```

```python
plt.figure(figsize=(19,17))
sns.heatmap(df.corr(),linewidths=.1, annot=True)
plt.yticks(rotation=0);
```

```python
X = df.drop(['class'], axis=1)
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred_dt = dt.predict(X_test)
print("Decision Tree Classifier report: \n\n", classification_report(y_test, y_pred_dt))
print("Test Accuracy: {}%".format(round(dt.score(X_test, y_test)*100, 2)))
```

```python
features = X.columns.values
featureI = dt.feature_importances_
sorted_idx = np.argsort(featureI)
plt.figure(figsize=(8,7))
plt.barh(range(len(sorted_idx)), featureI[sorted_idx])
plt.yticks(range(len(sorted_idx)), features[sorted_idx])
plt.xlabel('Importance')
plt.title('Feature importance')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(y_test, y_pred_dt)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
f, ax = plt.subplots(figsize =(10,10))
sns.heatmap(CM, annot = True, fmt = ".0f", cmap="Blues", ax=ax, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title('Mushroom Decision Tree Confusion Matrix')
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred_nb = nb.predict(X_test)
print("Naive Bayes Classifier report: \n\n", classification_report(y_test, y_pred_nb))
print("Test Accuracy: {}%".format(round(nb.score(X_test, y_test)*100, 2)))
```

```python
CM = confusion_matrix(y_test, y_pred_nb)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
f, ax = plt.subplots(figsize =(10, 10))
sns.heatmap(CM, annot = True, fmt = ".0f", cmap="Blues", ax=ax, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title('Mushroom Naive Bayes Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rforest = RandomForestClassifier(n_estimators=100, random_state=42)
rforest.fit(X_train, y_train)
```

```python
y_pred_rf = rforest.predict(X_test)
print("Random Forest Classifier report: \n\n", classification_report(y_test, y_pred_rf))
print("Test Accuracy: {}%".format(round(rforest.score(X_test, y_test)*100, 2)))
```

```python
CM = confusion_matrix(y_test, y_pred_rf)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
f, ax = plt.subplots(figsize =(10, 10))
sns.heatmap(CM, annot = True, fmt = ".0f", cmap="Blues", ax=ax, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title('Mushroom Random Forest Confusion Matrix');
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

best_Kvalue = 0
best_score = 0

for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    if knn.score(X_test, y_test) > best_score:
        best_score = knn.score(X_train, y_train)
        best_Kvalue = i
        
```

```python
y_pred_knn = knn.predict(X_test)
print("KNN Classifier report: \n\n", classification_report(y_test, y_pred_knn))
print("Test Accuracy: {}%".format(round(best_score*100,2)))
```

```python
cm = confusion_matrix(y_test, y_pred_knn)
x_axis_labels = ["Edible", "Poisonous"]
y_axis_labels = ["Edible", "Poisonous"]
f, ax = plt.subplots(figsize =(10, 10))
sns.heatmap(CM, annot = True, fmt = ".0f", cmap="Blues", ax=ax, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title('KNN Confusion Matrix')
plt.show()
```
